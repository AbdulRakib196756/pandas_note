<h1>marge</h1>

```python
tempareture_data=pd.DataFrame({
    'city':['york','washington','texas'],
    'temperature':[34,33,32],
    
})
tempareture_data
```

```python
humidity_data=pd.DataFrame({
    'city':['york','washington','texas'],
    'humidity':[88,60,70]
    
})
humidity_data
```

```python
df3=pd.merge(tempareture_data,humidity_data,on='city')
df3
# merge like intersection kebol common part add korbe 
```

```python
humidity_data=pd.DataFrame({
    'city':['york','washington','texas','dalas'],
    'humidity':[88,60,70,80]
    
})
humidity_data
```

```python
df3=pd.merge(tempareture_data,humidity_data,on='city')
df3
```

```python
# jehutu ekta row hocce na 
df3=pd.merge(tempareture_data,humidity_data,on='city',how='outer')
df3
```

```python
df3=pd.merge(tempareture_data,humidity_data,on='city',how='right',indicator=True)

df3
# 2nd dataframe and their common data 
```

```python
df2=pd.DataFrame({
    'city':['york','washingtion','texas'],
    'temperature':[21,22,23],
    'humidity':[64,66,77]
})
```

```python
df=pd.merge(df2,df3,on='city',suffixes=('_left','_right'))
df
```

<h1>melt</h1>
it is use for transform or reshape data 

```python
import pandas as pd
weather_data={
     'day':["monday",'tuesday','wednesday','thursday','friday','satureday','sunday'],
     'chicago':[28,24,33,35,33,34,32],
     'chennai':[75,78,76,67,76,78,67],
     'berlin':[42,43,42,41,45,44,46]

}
weather_data1=pd.DataFrame(weather_data)
weather_data1
```

```python
df1=pd.melt(weather_data1,id_vars=['day'],var_name='city',value_name='temperature')
df1
# reshape depends on day
```

```python

df1[df1['city']=='chicago']
```

<h1>crosstab<h1>

```python
import pandas as pd
df=pd.read_excel('survey.xlsx')
df
```

```python
pd.crosstab(df.nationality,df.handedness)
```

```python
pd.crosstab(df.sex,df.handedness,margins=True)
# margin is using for total
```

```python
# multiple coloumn
pd.crosstab(df.sex,[df.handedness,df.nationality])
```

```python
# if needed parcentage
pd.crosstab(df.sex,df.handedness,normalize='index')
```

```python
import numpy as np
pd.crosstab(df.sex,df.handedness,values=df.age,aggfunc=np.average)
# depends on average  value of age
```
